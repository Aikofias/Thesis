```python
import matplotlib.pyplot as plt

def plot_multiple_bars(df1, df2, df3, title1, title2, title3, ylabel):
    # Create a figure with 1 row and 3 columns (for 3 subplots)
    fig, ax = plt.subplots(1, 3, figsize=(18, 6))
    
    # Plot each DataFrame in a separate subplot
    df1.T.plot(kind="bar", ax=ax[0], legend=True)
    ax[0].set_title(title1)
    ax[0].set_ylabel(ylabel)
    ax[0].set_xlabel("Scenarios")
    ax[0].tick_params(axis='x', rotation=80)
    
    df2.T.plot(kind="bar", ax=ax[1], legend=True)
    ax[1].set_title(title2)
    ax[1].set_xlabel("Scenarios")
    ax[1].tick_params(axis='x', rotation=80)
    
    df3.T.plot(kind="bar", ax=ax[2], legend=True)
    ax[2].set_title(title3)
    ax[2].set_xlabel("Scenarios")
    ax[2].tick_params(axis='x', rotation=80)
    
    # Calculate the common y-axis limit from all three DataFrames
    all_data = pd.concat([df1, df2, df3])
    min_y = all_data.min().min()  # Minimum value across all
    max_y = all_data.max().max()  # Maximum value across all
    
    # Set the same y-axis limit for all subplots
    for axis in ax:
        axis.set_ylim(min_y - 0.1 * (max_y - min_y), max_y + 0.1 * (max_y - min_y))  # Adding some margin


    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()


def get_color(scenario_name):
    if scenario_name == "base":
        color = 'gray'
    if 'pv' in scenario_name.lower():
        # Extract kWp value and map it to a shade of blue
        kWp = float(scenario_name.lower().split('(')[1].split('kwp')[0])
        # Use a blue colormap, darker blue for higher kWp
        color = plt.cm.Blues(kWp / 10)  # Assuming the max kWp is 10 for scaling
    if 'ev' in scenario_name:
        # Extract kW value and map it to a shade of green
        kW = float(scenario_name.lower().split('(')[1].split('kw')[0])
        # Use a green colormap, darker green for higher kW
        color = plt.cm.Greens(kW / 10)  # Assuming the max kW is 10 for scaling
    if ('pv' in scenario_name.lower() and 'ev' in scenario_name.lower()):
        # Extract values for both kWp (for PV) and kW (for EV)
        kWp = float(scenario_name.lower().split('(')[1].split('kwp')[0]) #if 'pv' in scenario_name.lower() else 0
        kW = (scenario_name.lower().split('(')[-1].split('kw')[0]) #if 'ev' in scenario_name.lower() else 0

        if float(kW[0]) == 2:
            kW = float(2)
        elif float(kW[0]) == 7:
            kW = float(7)
        elif float(kW[0]) == 1:
            kW = float(11)
        combined = kWp + kW
        color = plt.cm.Reds(combined / 20)  # Assuming the max kW is 10 for scaling
 
    return color
#colors = [get_color(scenario) for scenario in combo.index]

def get_color_per_scenario(scenario_name):
    if scenario_name == "base":
        color = 'gray'
    if 'pv' in scenario_name.lower():
        color = "blue"
    if 'ev' in scenario_name:
        color = "green"
    if ('pv' in scenario_name.lower() and 'ev' in scenario_name.lower()):
        # Extract values for both kWp (for PV) and kW (for EV)
        kWp = float(scenario_name.lower().split('(')[1].split('kwp')[0]) #if 'pv' in scenario_name.lower() else 0
        kW = (scenario_name.lower().split('(')[-1].split('kw')[0]) #if 'ev' in scenario_name.lower() else 0
        g = float(0)
        b = float(0)
        if float(kW[0]) == 2:
            r = float(1)
            color=[r,g,b]
        elif float(kW[0]) == 7:
            r = float(0.6)
            color=[r,g,b]

        elif float(kW[0]) == 1:
            r = float(0.3)
            color=[r,g,b]

    return color

def plot_2_bars(df1, df2, title1, title2):
    # Create a figure with 1 row and 3 columns (for 3 subplots)
    fig, ax = plt.subplots(1, 2, figsize=(18, 8))
    
    # Plot each DataFrame in a separate subplot
    color_1 = [get_color(col) for col in df1.index]  # Get colors for df1 columns
    df1.T.plot(kind="bar", ax=ax[0], legend=True, color=color_1)
    ax[0].set_title(title1)
    #ax[0].set_ylabel("kW")
    ax[0].set_xlabel("Consumers")
    ax[0].tick_params(axis='x', rotation=80)
    
    color_2 = [get_color(col) for col in df2.index]  # Get colors for df1 columns
    df2.T.plot(kind="bar", ax=ax[1], legend=True, color =color_2)
    ax[1].set_title(title2)
    #ax[1].set_ylabel("kW")
    ax[1].set_xlabel("Consumers")
    ax[1].tick_params(axis='x', rotation=80)

    ax[0].set_xticklabels(["C1","C2","C3"], rotation=0)  # Custom labels for the first plot
    ax[1].set_xticklabels(["C1","C2","C3"], rotation=0)  # Custom labels for the second plot

    # Calculate the common y-axis limit from all three DataFrames
    all_data = pd.concat([df1, df2])
    min_y = all_data.min().min()  # Minimum value across all
    max_y = all_data.max().max()  # Maximum value across all
    
    # Set the same y-axis limit for all subplots
    for axis in ax:
        axis.set_ylim(min_y - 0.1 * (max_y - min_y), max_y + 0.1 * (max_y - min_y))  # Adding some margin

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```
