```python
import matplotlib.pyplot as plt

def plot_multiple_bars(df1, df2, df3, df4, df5, df6, title1, title2, title3, ylabel):
    fig, ax = plt.subplots(2, 3, figsize=(18, 10))
    
    # Plot each DataFrame in a separate subplot
    df1.T.plot(kind="bar", ax=ax[0,0], legend=True)
    ax[0,0].set_title(title1)
    ax[0,0].set_ylabel(ylabel)
    ax[0,0].set_xlabel("Scenarios")
    ax[0,0].tick_params(axis='x', rotation=80)
    
    df2.T.plot(kind="bar", ax=ax[0,1], legend=True)
    ax[0,1].set_title(title2)
    ax[0,1].set_xlabel("Scenarios")
    ax[0,1].tick_params(axis='x', rotation=80)
    
    df3.T.plot(kind="bar", ax=ax[0,2], legend=True)
    ax[0,2].set_title(title3)
    ax[0,2].set_xlabel("Scenarios")
    ax[0,2].tick_params(axis='x', rotation=80)
    
    df4.T.plot(kind="bar", ax=ax[1,0], legend=True)
    #ax[1,0].set_title(title1)
    ax[1,0].set_ylabel(ylabel)
    ax[1,0].set_xlabel("Scenarios")
    ax[1,0].tick_params(axis='x', rotation=80)
    
    df5.T.plot(kind="bar", ax=ax[1,1], legend=True)
    #ax[1,1].set_title(title2)
    ax[1,1].set_xlabel("Scenarios")
    ax[1,1].tick_params(axis='x', rotation=80)
    
    df6.T.plot(kind="bar", ax=ax[1,2], legend=True)
    #ax[1,2].set_title(title3)
    ax[1,2].set_xlabel("Scenarios")
    ax[1,2].tick_params(axis='x', rotation=80)

    # Calculate the common y-axis limit from all three DataFrames
    all_data = pd.concat([df1, df2, df3])
    all_data_2 = pd.concat([df4, df5, df6])
    all_data = pd.concat([all_data, all_data_2],axis=1)
    min_y = all_data.min().min()  # Minimum value across all    
    max_y = all_data.max().max()  # Maximum value across all
    # Set the same y-axis limit for all subplots

    for axis in ax.flatten():
        axis.set_ylim(min_y - 0.1 * (max_y - min_y), max_y + 0.1 * (max_y - min_y))  # Adding some margin


    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```
